```python
fig, axes = plt.subplots(nrows=1, ncols=2)

for ax in axes:
    ax.plot(x, y, 'r')
    ax.set_xlabel('x')
    ax.set_ylabel('y')
    ax.set_title('title')

fig.tight_layout()
```

```python
fig, ax = plt.subplots()

ax.plot(x, x**2, label="y = x**2")
ax.plot(x, x**3, label="y = x**3")
ax.legend(loc=2); # upper left corner
ax.set_xlabel('x')
ax.set_ylabel('y')
ax.set_title('title')
```

```python
# Гистограммы
from numpy import *
n = random.randn(100000)
fig, axes = plt.subplots(1, 2, figsize=(12, 4))
axes[0].hist(n)
axes[0].set_title("Default hystogram")
axes[0].set_xlim((min(n), max(n)))

axes[1].hist(n, cumulative=True, bins=50)
axes[1].set_title("Cumulative detailed hystogram")
axes[1].set_xlim((min(n), max(n)));
```

```python
# котировка магнита
from pylab import *
%matplotlib inline 

quotes = []
date = []
opn = []
in_fname = "MGNT_000101_170625.csv"

in_f = open(in_fname, 'r')
for line in in_f: 
    in_r = line.strip().split(';')
    quotes.append(line.strip().split(';'))
    date.append(in_r[2])
    opn.append(in_r[4])
```

```python
# построение скользящей средней
length = len(df_1)
n = 4
patt = df_1[4][-3:]
sma = [0]*len(df_1)
for i in range(n-1, length - 1):
    for k in range(n):
        #print(df_1[4][i-k])
        sma[i] = sma[i] + float(df_1[4][i-k])
    sma[i] = sma[i] / n
```
